#### Figure Size and DPI (tamaño de la figura y los pixeles de la misma)

```python
fig,axes = plt.subplots(nrows=2, ncols=1,figsize=(3,2), dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

#### Matplotlib part 3: Custumize figures

##### Plot Apparence

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', # color
        linewidth=0.5, # tamaño de linea
        alpha=0.5, # tranprancia
        linestyle = '--', # tipo de line
        marker='o', # tipo de puntos
        markersize = 20, # tamaño del punto
        markerfacecolor= 'yellow') # color dentro del marcador
# pudes poner RGB Hex color poniendo el #jdhf1445 / lw = linewidth
```

#### control over axis appearance / plot range

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=2, ls='--') 

ax.set_ylim([0,100]) # para manejar los limites
```
